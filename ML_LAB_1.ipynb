Write a program to create a numpy array using a python list . Display the array and print the shape and datatype of the array .



```python
import numpy as np
lst = [10, 20, 30, 40, 50]
arr = np.array(lst)

print("Array:", arr)
print("Shape:", arr.shape)
print("Data type:", arr.dtype)
```

Write a program to create 3X3 matrix of 0s , create 2X5 matrix of 1s . print both matrices and their shape.

```python
import numpy as np

zeros_matrix = np.zeros((3, 3))
ones_matrix = np.ones((2, 5))

print("3x3 Zero Matrix:\n", zeros_matrix)
print("Shape:", zeros_matrix.shape)

print("\n2x5 Ones Matrix:\n", ones_matrix)
print("Shape:", ones_matrix.shape)
```

Write a program to create a Numpy array of numbers from 1 to 12. Reshape the array into a 3X4 matrix. Display the original and reshape array along with their shapes.

```python
import numpy as np

arr = np.arange(1, 13)
reshaped_arr = arr.reshape(3, 4)

print("Original Array:", arr)
print("Original Shape:", arr.shape)

print("\nReshaped Array:\n", reshaped_arr)
print("Reshaped Shape:", reshaped_arr.shape)
```

Write a program to create Numpy array of 10 random integers . Find min , max , mean and sum of elements of the arrays.

```python
import numpy as np

arr = np.random.randint(1, 100, 10)

print("Array:", arr)
print("Minimum:", arr.min())
print("Maximum:", arr.max())
print("Mean:", arr.mean())
print("Sum:", arr.sum())
```

Write a program to read a csv file into a data frame and display the statistical summary using describe method. Interpret mean, standard deviation, min and max values

```python
import pandas as pd

df = pd.read_csv("DataState.csv")

print("Statistical Summary using describe():\n")
print(df.describe())

print("\nMean of each numerical column:\n")
print(df.mean(numeric_only=True))

print("\nStandard Deviation of each numerical column:\n")
print(df.std(numeric_only=True))

print("\nMinimum value of each numerical column:\n")
print(df.min(numeric_only=True))

print("\nMaximum value of each numerical column:\n")
print(df.max(numeric_only=True))
```
