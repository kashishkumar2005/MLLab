```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("DataState.csv")

years = [
    'Total Number of Persons Injured in Road Accidents on State Highways during 2019',
    'Total Number of Persons Injured in Road Accidents on State Highways during 2020',
    'Total Number of Persons Injured in Road Accidents on State Highways during 2021',
    'Total Number of Persons Injured in Road Accidents on State Highways during 2022',
    'Total Number of Persons Injured in Road Accidents on State Highways during 2023'
]

total_injuries = df[years].sum()

plt.plot([2019, 2020, 2021, 2022, 2023], total_injuries)
plt.xlabel("Year")
plt.ylabel("Total Injuries")
plt.title("Trend of Road Accident Injuries Over Years")
plt.show()

plt.bar(df['States/UTs'], df[years[4]])
plt.xticks(rotation=90)
plt.xlabel("States/UTs")
plt.ylabel("Injuries in 2023")
plt.title("State-wise Road Accident Injuries (2023)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("DataState.csv")

col = 'Total Number of Persons Injured in Road Accidents on State Highways during 2023'

data = pd.to_numeric(df[col], errors='coerce')
data = data.dropna()

plt.figure(figsize=(10, 4))
plt.boxplot(data, vert=False)
plt.xlabel("Number of Injuries")
plt.title("Boxplot of Road Accident Injuries (2023)")
plt.show()
```
