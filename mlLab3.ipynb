```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

print(df.head())

print(df.dtypes)

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print(numerical_cols)
print(categorical_cols)

print(df[numerical_cols].mean())
print(df[numerical_cols].median())
print(df[numerical_cols].mode())
print(df[numerical_cols].min())
print(df[numerical_cols].max())
print(df[numerical_cols].sum())
print(df[numerical_cols].var())
print(df[numerical_cols].std())
```

```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

numerical_cols = df.select_dtypes(include=['int64', 'float64'])

Q1 = numerical_cols.quantile(0.25)
Q2 = numerical_cols.quantile(0.50)
Q3 = numerical_cols.quantile(0.75)

print(Q1)
print(Q2)
print(Q3)
```

```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

numerical_cols = df.select_dtypes(include=['int64', 'float64'])

print(numerical_cols.corr())

print(numerical_cols.cov())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

numerical_cols = df.select_dtypes(include=['int64', 'float64'])

numerical_cols.hist(figsize=(8,6))
plt.show()

plt.boxplot(numerical_cols.values, labels=numerical_cols.columns)
plt.show()
```
